# Prints sum of male and female in respective year

```python
pd.pivot_table(df, values='births', index=['year'],columns=['gender'],aggfunc=np.sum)
```

pd.pivot_table(df,values='births',index=['year'],columns=['gender'],margins=True,aggfunc=sum)

# Print sum of Female and males in respective year in ascending order and use of margin as new column All is added

```python
pd.pivot_table(df,values='births',index=['year'],columns=['gender'],margins=True,aggfunc=sum).sort_values('All')
```

# Print last 5 values using tail mehod

```python
pd.pivot_table(df,values='births',index=['year'],columns=['gender'],margins=True,aggfunc=sum).sort_values('All').tail(5)
```
